### use word2vec model from tohoku university

```python
# download from http://www.cl.ecei.tohoku.ac.jp/~m-suzuki/jawiki_vector/
model = KeyedVectors.load_word2vec_format("entity_vector/entity_vector.model.bin", binary=True)
```

### difine waaktigaki

```python
# Defined by SlpothLib
slothlib_path = 'http://svn.sourceforge.jp/svnroot/slothlib/CSharp/Version1/SlothLib/NLP/Filter/StopWord/word/Japanese.txt'
slothlib_file = urllib.request.urlopen(slothlib_path)
slothlib_stopwords = [line.decode("utf-8").strip() for line in slothlib_file]
stopwords = [ss for ss in slothlib_stopwords if not ss==u'']
# stopwords = stopwords + ["もらい","もらう","られる","し", "い", "する", "さ", "れる", "なり", "しまい", "しまっ", "れ", "あり", "なろ", "ある", "せる", "なる","の", "お", "いる", "できる", "でき", "さら", "られ", "きっ", "おり", "か月", "わかっ"]

def wakatigaki(text):
    
    """
    Get word and class tuples list.
    """
    # format
    text=re.sub(r'[!-~]', "", text) #半角記号,数字,英字
    text = re.sub(r'[、。︰-＠]', "", text) #全角記号
    text = re.sub('\n', " ", text) #改行文字

    # Execute class analysis
    tagger = MeCab.Tagger('-Ochasen')
    result = tagger.parse(text)

    # Extract word and class
    words = []
    for input_value in result.split("\n"):
        # check 
        condition1 = input_value.find("名詞") > 0
        condition2 = (input_value.find("動詞") > 0 and input_value.find("助動詞") < 0)
        if (condition1 or condition2):
            words.append(input_value.split("\t")[0])
                        
    words_re = []
    for w in words:
        if not w in stopwords:
            words_re.append(w)
    
    return words_re
```

### wakatigaki for all contents

```python
def apply_wakatigaki(input_target):
    Awvl = [None] * len(input_target)
    for (i,value_1) in enumerate(input_target):
        wl = wakatigaki(value_1)
        wvl = [None] * len(wl)
        for (j,x) in enumerate(wl):
            try:
                wvl[j] = np.array(model[x])
            except:
                pass
        wvl2 = [x for x in wvl if x is not None]
        Awvl[i] = pd.DataFrame(wvl2).mean().values
    rAwvl = pd.DataFrame(Awvl).values
    return rAwvl
```

# Question.1
### 困難に向き合い、乗り越え、実現してきた経験を教えてください。（300字以内）

```python
target_1 = data["3"]
target_1 = target_1.drop_duplicates()
target_1.dropna(inplace=True)
NAwvl_1 = apply_wakatigaki(target_1)
```

### Kmeans

```python
nc = 10
kmeans = KMeans(n_clusters=nc, random_state=0)
kmeans.fit(NAwvl_1)
```

```python
target_1_pd = target_1.to_frame()
target_1_pd["label"] = kmeans.labels_
for i in range(nc):
    print("Cluster ", i)
    display(target_1_pd[target_1_pd["label"] == i])
```
